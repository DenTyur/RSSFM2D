```python
import numpy as np
from matplotlib import cm
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import os
basedir = os.path.abspath(os.getcwd())
src_dir = os.path.abspath(os.path.join(basedir, '..'))
src_dir
```

```python
x = np.load(src_dir + "/arrays_saved/x0.npy")
y = np.load(src_dir + "/arrays_saved/x1.npy")
atomic_potential = np.load(src_dir + "/arrays_saved/atomic_potential.npy")
psi_initial = np.load(src_dir + "/arrays_saved/psi_initial.npy")
```

```python
X, Y = np.meshgrid(x, y, indexing='ij')
```

```python
fig, ax = plt.subplots(1, 1, figsize=(5, 5))
ax.set(aspect="equal", title = "Atomic and Absorbing potentials")
b = ax.pcolormesh(X, Y, np.abs(atomic_potential), cmap=cm.jet, shading="auto")
cb = pl.colorbar(b, ax=ax)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(5, 5))
ax.set(aspect="equal", title = "Initial Wave Function")
b = ax.pcolormesh(X, Y, np.abs(psi_initial), cmap=cm.jet, shading="auto")
cb = pl.colorbar(b, ax=ax)
```
